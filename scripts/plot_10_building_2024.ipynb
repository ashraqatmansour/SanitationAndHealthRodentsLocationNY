```python
df = pd.read_csv("../data/building_count.csv")
top10 = df.nlargest(10, "Count")
plt.figure(figsize=(10,6))
plt.barh(top10["Building Type"], top10["Count"], color="skyblue")
plt.xlabel("Rodent Complaint Count (2024)")
plt.ylabel("Building Type")
plt.title("Top 10 Building Types for Rodent Complaints in 2024")
plt.gca().invert_yaxis()  # Makes the largest bar appear at the top
plt.tight_layout()
plt.show()
```
